```python
for j in dd1.index:
    v = dd1.loc[j,'Variable']
    # m = dd.loc[j,'Element Code']
    i = dd1.loc[j,'Item Code']
    # u = dd.loc[j,'Units']
    if dd1.isnull().loc[j,'Formula']:
        f = dd1.loc[j,'Formula']
    else:
        f = "val" + str(dd1.loc[j,'Formula'])
    if dd.isnull().loc[j,'Decimal Places']:
        d = dd1.loc[j,'Decimal Places']
    else:
        d = int(dd1.loc[j,'Decimal Places'])
    print(v,f,d)
    dt = df[df['Item Code'] == i]
    dt = dt[['Area','Year','Value']]
    dt = dt.rename(columns={'Value':'val'})
    if dd1.isnull().loc[j,'Formula']:
        dt = dt
    else:
        dt.val = dt.eval(f)
    if dd1.isnull().loc[j,'Decimal Places']:
        dt = dt
    else:
        dt.val = round(dt.val, d)
    
    dt = country.merge(dt,
                        on = 'Area',
                        how = 'left')
    dt.drop(columns='Area',
            inplace=True)

    dt = dt.groupby(['Country', 'FIPS_CODE','Year']).sum(min_count=1).reset_index()
    first_last = dt.sort_values(['Country','Year']).groupby(dt['Country'])['val'] \
        .agg(['first', 'last'])
    first_last = first_last.rename(columns={"first": "Earliest", "last": "MostRecent"})
    dt.Year = dt.Year.astype('Int64')
    dt = dt.pivot_table(index = ['Country', 'FIPS_CODE'],
                    columns='Year',
                    values = 'val').reset_index()
    dt = dt.merge(first_last,
                        on = 'Country',
                        how = 'left'
                        )
    dt = country.merge(dt,
                        on = ['Country','FIPS_CODE'],
                        how = 'left')
    dt.drop(columns='Area', inplace = True)
    dt.dropna(axis = 0, how = 'all', inplace=True)
    dt = dt.drop_duplicates(ignore_index=True)
    dt = dt.sort_values(by='Country')
    dt = dt.set_index(['Country','FIPS_CODE'])
    if dd1.isnull().loc[j,'Decimal Places']:
        dt = dt
    else:
        dt = dt.apply(lambda x: pd.Series.round(x, d))
    dt.to_csv(f'IFs Import/{v}.csv',
                #  index=False
                 )
dt
```

## LandOther%

```python

df = pd.read_csv('IFs Import/full/LandTotal.csv')
dt = pd.read_csv('IFs Import/full/LandOther.csv')

df = df.drop(columns=['Earliest','MostRecent'])
df = df.melt(id_vars= ["Country","FIPS_CODE"],var_name="Year",value_name="total")

dt = dt.drop(columns=['Earliest','MostRecent'])
dt = dt.melt(id_vars= ["Country","FIPS_CODE"],var_name="Year",value_name="other")

dd = df.merge(dt, on = ['Country','FIPS_CODE','Year'])
dd['val'] = round(dd['other']/dd['total']*100,3)



n = 'Country'
y = 'Year'
first_last = dd.sort_values([n,y]).groupby(dd[n])['val'] \
        .agg(['first', 'last']).reset_index()
first_last = first_last.rename(columns={"first": "Earliest", "last": "MostRecent"})

dd = dd.pivot_table(index = n,
                columns= y,
                values = 'val').reset_index()
dd = dd.merge(first_last,
            on = n,
            how = 'left')
dd = country.merge(dd,
                on = n,
                how = 'left')
dd = dd.drop(columns= 'Area')
dd = dd.drop_duplicates()
dd.to_csv('IFs Import/full/LandOther%.csv', index = False)
```

## LandPasture%

```python

df = pd.read_csv('IFs Import/full/LandTotal.csv')
dt = pd.read_csv('IFs Import/full/LandGrazing.csv')

df = df.drop(columns=['Earliest','MostRecent'])
df = df.melt(id_vars= ["Country","FIPS_CODE"],var_name="Year",value_name="total")

dt = dt.drop(columns=['Earliest','MostRecent'])
dt = dt.melt(id_vars= ["Country","FIPS_CODE"],var_name="Year",value_name="pasture")

dd = df.merge(dt, on = ['Country','FIPS_CODE','Year'])
dd['val'] = round(dd['pasture']/dd['total']*100,3)



n = 'Country'
y = 'Year'
first_last = dd.sort_values([n,y]).groupby(dd[n])['val'] \
        .agg(['first', 'last']).reset_index()
first_last = first_last.rename(columns={"first": "Earliest", "last": "MostRecent"})

dd = dd.pivot_table(index = n,
                columns= y,
                values = 'val').reset_index()
dd = dd.merge(first_last,
            on = n,
            how = 'left')
dd = country.merge(dd,
                on = n,
                how = 'left')
dd = dd.drop(columns= 'Area')
dd = dd.drop_duplicates()
dd.to_csv('IFs Import/full/LandPasture%.csv', index = False)

```
